```python
"""
Program: Gradient Descent to find the local minima of a function
Example: Find the local minima of y = (x + 3)^2 starting from x = 2

Reference: https://en.wikipedia.org/wiki/Gradient_descent
"""

import matplotlib.pyplot as plt

# Function to minimize
def f(x: float) -> float:
    return (x + 3) ** 2

# Derivative of the function
def df_dx(x: float) -> float:
    return 2 * (x + 3)

# List to store (x, f(x)) values for visualization
curve = []

# User input
x_0 = float(input("Enter initial value for 'x' (x_0): "))
n = int(input("Enter max. number of steps: "))
step_size = float(input("Enter step size (learning rate): "))

# Starting point
curve.append((x_0, f(x_0)))
x_t = x_0

# Gradient Descent Algorithm
for i in range(n):
    grad = df_dx(x_t)
    x_t = x_t - step_size * grad
    print(f"Step {i + 1}: x = {x_t:.4f}, f(x) = {f(x_t):.4f}, gradient = {grad:.4f}")
    curve.append((x_t, f(x_t)))

# Final Result
print("\n✅ Local minimum found:")
print(f"x = {x_t:.4f}")
print(f"f(x) = {f(x_t):.4f}")

# Plot convergence
x, y = zip(*curve)
plt.plot(x, y, marker='o', color='b')
plt.title("Gradient Descent Convergence")
plt.xlabel("x values")
plt.ylabel("f(x) values")
plt.grid(True)
plt.show()
```
